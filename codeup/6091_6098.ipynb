```python
#6091
a,b,c = map(int,input().split())

for i in range(max(a,b,c),(a*b*c+1)):
    if i%a==0 and i%b==0 and i%c==0:
        print(i)
        break
```

```python
#6092
a=[]

for i in range(24):
    a.append(0) # 0으로 리스트 초기화
n = int(input()) # 출석 번호를 부른 횟수

b = input().split() # 무작위로 부른 n개의 번호(1~23)

for i in range(n):
    a[int(b[i])]+=1 #불려진 번호 횟수 세기
for i in range(1,len(a)):#(1~23)이므로 1부터 세기
    print(a[i],end=' ') #불려진 번호 횟수 출력
```

```python
#6093
n = int(input())
a = input().split()
for i in range(n):
    a[i]=int(a[i])

for i in range(n-1,-1,-1):
    print(a[i],end=' ')
```
